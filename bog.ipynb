```python
import codecs
import time
```

```python
fname = 'untitled.txt'

words = []
```

```python
fsent = codecs.open(fname, "r", encoding='utf8')
```

```python
for line in fsent:
    line = line.strip()
    linelist = line.split()
    if ((len(linelist) == 1) and ("ctx" in linelist[0])):
        continue  # To skip the sentence count marker line.
    elif (len(linelist) == 0):
         continue
    else:
        size = len(linelist)
        for j in range(0, size):  # It will iterate from index 0 to index (size-1).
            item = linelist[j].strip()
            senselist = item.split("#")
            midword = senselist[0]
            words.append(midword)
```
